"flag"   , . flag = -15, ,     .   . <a href=http://www.desy.de/~mpyflo/Astra_manual/Astra-Manual_V3.2.pdf></a> (. 3).

```python
#          :

def read_track(fname):
    df = pd.read_csv(fname, header=None, delim_whitespace=True, names=cols, dtype='float32')
    
    df = df[df.flag !=-15] #   
    
    df['x'] = df['x']*1e3 # mm
    df['y'] = df['y']*1e3 # mm
    
    df['px'] = df['px']/1e6 # MeV/c
    df['py'] = df['py']/1e6 # MeV/c
    
    #   
    df0 = df.head(1)
    df  = df.drop(df0.index)
    
    z0  = df0.z.values[0]
    pz0 = df0.pz.values[0]
    
    #  pz  z:
    
    df['z'] = z0 + df['z'] # m
    df['pz'] = (pz0 + df['pz'])/1e6 # MeV/c
    
    #        100000:
    #df = df.sample(n=100000)
    
    return df
```

```python
df = read_track(track)
```
